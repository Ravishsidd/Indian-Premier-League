```python
# Getting the frequency of Top 10 player with most Man of the Match:
list(ipl["player_of_match"].value_counts()[0:10].keys())
```

```python
 # Bar plot of Top 10 player with most Man of the match:
plt.figure(figsize=(10,5))
plt.bar(list(ipl['player_of_match'].value_counts()[0:10].keys()),list(ipl['player_of_match'].value_counts()[0:10]),color='green')
plt.title('Most_Man_of_the_Matches_awards')
plt.xlabel('Players')
plt.ylabel('No_of_Man_0f_the_Matches')
plt.show()
```

```python
# Getting the frequency of Result_column [Result_Column is a categorical_column]:
ipl['result'].value_counts()
```

```python
# Extracting the Records where a Team won by First Batting (win_by_runs):
Batting_first=ipl[ipl['win_by_runs']!=0]
```

```python
Batting_first.head()
```

```python
Batting_first.tail()
```

```python
# Making Histogram of Team won by Batting_first (win_by_runs):
plt.figure(figsize=(10,5))
plt.hist(Batting_first['win_by_runs'],color='brown')
plt.title('Distribution_of_Runs')
plt.xlabel('Runs')
plt.ylabel('Matches')
plt.show()
```

```python
# Getting the Frequency of win w.r.t each Team after Batting First:
Batting_first['winner'].value_counts()
```

```python
# Histogram of Teams with most_wins after Batting_first:
plt.figure(figsize=(35,20))
plt.hist(Batting_first['winner'],color='orange')
plt.title('Distribution_of_Winning_Teams_by_Batting_first')
plt.xlabel('No_of_Teams')
plt.ylabel('No_of_Matches')
plt.show()
```

```python
# Getting the Frequency of win w.r.t each Team after Batting First:
Batting_first['winner'].value_counts()
```

```python
# Barplot of Teams with most_wins after Batting_first:
plt.figure(figsize=(30,9))
plt.bar(list(Batting_first['winner'].value_counts().keys()),list(Batting_first['winner'].value_counts()),color='purple')
plt.title('Most_win_Team_afterr_Batting_First')
plt.xlabel('N0_of_Teams')
plt.ylabel('No_of_Match_Wins')
plt.show()
```

```python
# Pie_chart of Teams with most_wins after Batting_first:
plt.figure(figsize=(40,10))
plt.pie(list(Batting_first['winner'].value_counts()),labels=list(Batting_first['winner'].value_counts().keys()),autopct='%0.1f%%')
plt.show()
```
